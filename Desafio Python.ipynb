```python
import pandas as pd
import re
import numpy as np
```

```python
arquivo = open('Base_Teste_coordenação_de_vendas.txt', 'r', encoding="latin-1")
```

```python
#pegar a primeira linha para criar as colunas no dataframe
colunas = arquivo.readline()
```

```python
#recebe uma linha como parametro e retorna a linha sem tabulacoes e pulo de linhas
#retorna uma lista
def tratamentoDados(linha):
    #removendo tabulações e pulo de linha
    dados= linha.replace('\n','')

    #split na string para ter a lista
    dados = re.split('\t', linha)
    dados.pop()
    
    return dados
```

```python
colunas = tratamentoDados(colunas)
```

```python
dataframe = pd.DataFrame(columns=colunas)
```

```python
for linha in arquivo:
    #transforma a linha em uma lista
    dados=tratamentoDados(linha)
    
    #adicionando ao dataframe
    dataframe.loc[len(dataframe)]=dados
    
arquivo.close()
```

```python
dataframe.to_csv('dados.csv')
```
